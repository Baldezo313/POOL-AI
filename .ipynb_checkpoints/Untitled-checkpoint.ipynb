```python
url = 'https://cadastre.data.gouv.fr/bundler/cadastre-etalab/departements/83/geojson/'
```

```python
def download_file(url_file, save_name = False) :
    '''
    Download and save file from an url
    '''
    if save_name == False :
        save_name = url_file.split('/')[-1]
        
    r = requests.get(url_file, stream = True)
    with open(save_name, 'wb') as f:
        for chunk in r.iter_content():
            f.write(chunk)
            
```

```python
def parcelle_to_code(p) :
    if p :
        r = p[0:6]
    else :
        r = p
    return r

def parcelle_to_lot(p) :
    if p :
        #print(p)
        r = p[-5::]
       #print(r)
    else :
        r = p
    return r
```

```python
subdivisions_fiscales['code'] = subdivisions_fiscales.apply(lambda x : parcelle_to_code(x['parcelle']), axis = 1 )
subdivisions_fiscales['lot'] = subdivisions_fiscales.apply(lambda x : parcelle_to_lot(x['parcelle']), axis = 1 )

subdivisions_fiscales.head()
```
